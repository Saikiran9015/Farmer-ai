```python

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import os
```

```python
DATA_PATH = r"C:\Users\abbus\Desktop\mini project\human_chat.txt"
```

```python
dialogs = []
with open(DATA_PATH, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()

user_msg, ai_msg = None, None
for line in lines:
    line = line.strip()
    if line.startswith("Farmer:"):
        user_msg = line[len("Farmer:"):].strip()
    elif line.startswith("AI:"):
        ai_msg = line[len("AI:"):].strip()
    elif line == "":
        if user_msg and ai_msg:
            dialogs.append((user_msg, ai_msg))
        user_msg, ai_msg = None, None
```

```python
if user_msg and ai_msg:
    dialogs.append((user_msg, ai_msg))

print(f"Loaded {len(dialogs)} dialogues")

user_texts = [u for (u, a) in dialogs]
ai_texts   = [a for (u, a) in dialogs]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
user_embeddings = model.encode(user_texts)
```

```python
def get_response(query, threshold=0.6):
    query_emb = model.encode([query])[0]
    sims = cosine_similarity([query_emb], user_embeddings)[0]
    best_idx = np.argmax(sims)
    best_score = sims[best_idx]
    if best_score >= threshold:
        return ai_texts[best_idx]
    else:
        return ("Sorry — I don't have an appropriate answer to that. "
                "Please ask about mandi price, crop, or market.")
```

```python
print(get_response("What is the current rice price at Hyderabad Mandi?"))
```

```python
print("=== Farmer-AI Chatbot ===")
print("Type 'exit' to end.")
while True:
    usr = input("Farmer: ")
    if usr.strip().lower() in ["exit", "quit"]:
        print("AI: Goodbye — take care!")
        break
    resp = get_response(usr)
    print("AI:", resp)
```

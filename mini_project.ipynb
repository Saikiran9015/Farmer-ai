## Text Preprocessing and Vectorization

### Subtask:
Apply TF-IDF (Term Frequency-Inverse Document Frequency) vectorization to convert the textual user messages into numerical features.


**Reasoning**:
The subtask requires applying TF-IDF vectorization to the text data. This involves importing the necessary class, instantiating it, fitting it to the input features (X), and then transforming X to create numerical features.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
X_vectorized = tfidf_vectorizer.fit_transform(X)

print("Shape of X_vectorized:", X_vectorized.shape)
print("Type of X_vectorized:", type(X_vectorized))
```

## Train NLP Classification Model

### Subtask:
Split the vectorized data into training and testing sets. Then, train a suitable classification model (such as a Logistic Regression or a Support Vector Machine) on the training data to predict intent.


**Reasoning**:
The next step is to prepare the data for model training by splitting it into training and testing sets, and then to train a Logistic Regression model on the training data. This is a fundamental step in machine learning to evaluate model performance on unseen data.



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# 1. Split the vectorized data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

# 2. Instantiate a Logistic Regression model
logistic_model = LogisticRegression(max_iter=1000, random_state=42)

# 3. Train the Logistic Regression model
logistic_model.fit(X_train, y_train)

print("\nLogistic Regression model trained successfully.")
```
